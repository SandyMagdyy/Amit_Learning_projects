```python
import numpy as np                  # For mathematical operations
import pandas as pd                 # For working with data or .csv file
import matplotlib.pyplot as plt     # For general visualizations
import seaborn as sns               # Much better visualizations
```

### Data Exploration and Cleaning

```python
#importing data
df= pd.read_csv("Breast Cancer Wisconsin (Diagnostic).csv")
df.head()
```

```python
# numbers of rows and columns
df.shape
```

```python
# knowing more information about the data
df.info()
```

```python
# check the nulls
df.isnull().sum()
```

```python
# deleting un wanted rows
df.drop(["Unnamed: 32", "id"] , axis=1 , inplace=True)
```

```python
#check number of rows and columns
df.shape
```

```python
# counting the Benign and Malignant
df["diagnosis"].value_counts()
```

```python
# percentage of Benign and Malignant       (Bengin--->62%, Malignant---->37%)
df["diagnosis"].value_counts(normalize=True)
```

```python
# change data type from categorical to numeric values In order to conduct the analysis easily (Benign=0, Malignant=1)
df["diagnosis"]=df["diagnosis"].map({"M":1, "B":0})
```

```python
# checking
df
```

```python
# summary statistics to know more info
df.describe()
```

### Data visualization

```python
# visualize the count of tumor type Bengin(0) and Malignant(1)
sns.countplot(x=df["diagnosis"])
plt.title("count of Tumor type Bengin(0) and Malignant(1)")
plt.show()
```

```python
# percentage of Tumor type (Begnign=0, Malignant=1)
df["diagnosis"].value_counts(normalize=True).plot(kind="pie",autopct='%1.0f%%')
plt.title("percentage of Tumor type Benign(0) and Malignant(1)")
plt.show()
```

##### - we can observe from the heatmap that The blue correspond to malignant diagnosis and red to benign. for example the more raduis means the more chance to be malignant and the less smoothness means the more chance to be bengin

```python
# correlation of features by diagnosis (we can observe that The blue correspond to malignant diagnosis and red to benign. )
plt.figure(figsize=(20,20))
plt.title("Breast cancer feature correlation by diagnosis", fontsize=40)
sns.heatmap(df.corr(), annot=True, cmap="RdBu")
plt.show()
```
